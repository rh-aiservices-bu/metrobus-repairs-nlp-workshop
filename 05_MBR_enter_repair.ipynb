```python
#my_text is the repair text that a bus driver enters at the end of the shift, if there is an issue with his vehicle
my_text = 'car will not stop'     

#my_route is the route that was created in your OpenShift dedicated project
my_route = 'http://metrobus-repairs-nlp-workshop-metrobus-repairs.apps.integrations.fb55.p1.openshiftapps.com/'
 
#my_route = 'http://metrobus-repairs-nlp-workshop-metrobus-repairs3.apps.integrations.fb55.p1.openshiftapps.com/'
```

```python
import requests
import json
from json import dumps

content = {"text": my_text}
json_data = dumps(content)
print(json_data)
headers = {"Content-Type" : "application/json"}

r = requests.post(my_route + '/prediction', data=json_data, headers=headers)

print(r.content)

```
